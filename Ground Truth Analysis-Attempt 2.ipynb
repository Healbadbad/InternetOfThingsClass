```python
import matplotlib.pyplot as plt
import numpy as np
from iotdata import IOTData
devices = ["Ground_Truth_Treadmill1", "Ground_Truth_Treadmill2", "Ground_Truth_Treadmill3", "Ground_Truth_Treadmill5"]
device = IOTData(devices, 'Ground_Truth_data\\03_27_16\\')

def plotarray(data):  
    length = len(data)
    for k in range(length):
        plt.plot(range(len(data[k])), data[k])
```

```python
for k in range(len(devices)):
    figsize(17, 6)
    figure(k)
    device.initdata(devices[k])
    length = device.getLength()
    data = device.getSelection(0, length)
    plotarray(data [11:14])
```

```python
device.initdata(devices[1])
data = device.getSelection(120300, 500000)
# plotarray(data[1:4])
plt.plot(data[0], data[11])
plt.plot(data[0], data[12])
plt.plot(data[0], data[13])
```

```python
device.initdata(devices[1])
data = device.getSelection(170000, 250000)
# plotarray(data[1:4])
plt.plot(data[0], data[3])
```
